```python
# Dependencies and Setup
%matplotlib inline
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Hide warning messages in notebook
import warnings
warnings.filterwarnings('ignore')

# File to Load (Remember to Change These)
mouse_drug_data_to_load = "data/mouse_drug_data.csv"
clinical_trial_data_to_load = "data/clinicaltrial_data.csv"

# Read the Mouse and Drug Data and the Clinical Trial Data

mouseData = pd.read_csv(mouse_drug_data_to_load)
clinicalTrialData = pd.read_csv(clinical_trial_data_to_load)

# Combine the data into a single dataset

mouse_clinical_Data = pd.merge(clinicalTrialData,mouseData,on='Mouse ID')

# Display the data table for preview

mouse_clinical_Data.head()
```

## Tumor Response to Treatment

```python
# Store the Mean Tumor Volume Data Grouped by Drug and Timepoint 
#tumorByDrugTime = mouse_clinical_Data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].agg(['mean'])

tumorByDrugTime = mouse_clinical_Data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].agg(['mean'])

# Convert to DataFrame
tumorByDrugTime.reset_index(inplace = True)

tumorByDrugTime.rename(columns = {'mean':'Tumor Volume (mm3)'}, inplace=True)

# Preview DataFrame
tumorByDrugTime.head()
```

```python
# Store the Standard Error of Tumor Volumes Grouped by Drug and Timepoint
tumorVol_SE = mouse_clinical_Data.groupby(['Drug','Timepoint'])['Tumor Volume (mm3)'].agg('sem')

# Convert to DataFrame
tumorVol_SE = pd.DataFrame(tumorVol_SE)
tumorVol_SE.reset_index(inplace = True)

# Preview DataFrame
tumorVol_SE.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
tumorByDrugTime_Formatted = tumorByDrugTime.pivot(index = 'Timepoint', columns = 'Drug', values = 'Tumor Volume (mm3)')

# Preview that Reformatting worked
tumorByDrugTime_Formatted.head()
```

```python
# Get the Timepoint Value which is the index of tumorByDrugTime dataframe into x_axis
x_axis_volume = tumorByDrugTime_Formatted.index.values

# Generate the Plot (with Error Bars)
plt.plot(x_axis_volume,tumorByDrugTime_Formatted['Capomulin'],label='Capomulin',\
         color ='red', marker = 'o', linestyle ='dashed',linewidth = 0.5, markersize = 5)
plt.plot(x_axis_volume,tumorByDrugTime_Formatted['Infubinol'],label='Infubinol',\
         color ='blue', marker = '^', linestyle ='dashed',linewidth = 0.5, markersize = 5)
plt.plot(x_axis_volume,tumorByDrugTime_Formatted['Ketapril'],label='Ketapril',\
         color ='green', marker = 's', linestyle ='dashed',linewidth = 0.5, markersize = 5)
plt.plot(x_axis_volume,tumorByDrugTime_Formatted['Placebo'],label='Placebo',\
         color ='black', marker = 'd', linestyle ='dashed',linewidth = 0.5, markersize = 5)

plt.legend(loc = 'best')
plt.xlabel('Time (Days)')
plt.ylabel('Tumor Volume (mm3)')
plt.title('Tumor Response to Treatment')
plt.grid()

plt.ylim(min(tumorByDrugTime_Formatted['Capomulin'])-3,max(tumorByDrugTime_Formatted['Ketapril'])+3)
#plt.show()

# Save the Figure
plt.savefig('./Images/1_Tumor Response to Treatment.png')
```

## Metastatic Response to Treatment

```python
# Store the Mean Met. Site Data Grouped by Drug and Timepoint 
metastatic_Summay = mouse_clinical_Data.groupby(['Drug','Timepoint'])['Metastatic Sites'].agg('mean')

# Convert to DataFrame
metastatic_Summay = pd.DataFrame(metastatic_Summay)

# Preview DataFrame
metastatic_Summay.head()
```

```python
# Store the Standard Error associated with Met. Sites Grouped by Drug and Timepoint 
metSites_SE = mouse_clinical_Data.groupby(['Drug','Timepoint'])['Metastatic Sites'].agg('sem')

# Convert to DataFrame
metSites_SE = pd.DataFrame(metSites_SE)

# Preview DataFrame
metSites_SE.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
metastatic_Summay.reset_index(inplace=True)
metastatic_Summay_Formatted = metastatic_Summay.pivot(index='Timepoint',columns='Drug',values='Metastatic Sites')

# Preview that Reformatting worked
metastatic_Summay_Formatted.head()
```

```python
x_axis_metastatis = metastatic_Summay_Formatted.index.values

# Generate the Plot (with Error Bars)
plt.plot(x_axis_metastatis,metastatic_Summay_Formatted['Capomulin'],label='Capomulin',\
         color ='red', marker = 'o', linestyle ='dashed',linewidth = 0.5, markersize = 5)
plt.plot(x_axis_metastatis,metastatic_Summay_Formatted['Infubinol'],label='Infubinol',\
         color ='blue', marker = '^', linestyle ='dashed',linewidth = 0.5, markersize = 5)
plt.plot(x_axis_metastatis,metastatic_Summay_Formatted['Ketapril'],label='Ketapril',\
         color ='green', marker = 's', linestyle ='dashed',linewidth = 0.5, markersize = 5)
plt.plot(x_axis_metastatis,metastatic_Summay_Formatted['Placebo'],label='Placebo',\
         color ='black', marker = 'd', linestyle ='dashed',linewidth = 0.5, markersize = 5)


plt.legend(loc = 'best')
plt.xlabel('Time (Days)')
plt.ylabel('Met. Sites')
plt.title('Metastatic Spread During Treatment')
plt.grid()
plt.ylim(-0.2,3.8)

# Save the Figure
plt.savefig('./Images/2_Metastic Spread During Treatment.png')

# Show the Figure
plt.show()
```

## Survival Rates

```python
# Store the Count of Mice Grouped by Drug and Timepoint (W can pass any metric)

miceByDrugTimepoint = mouse_clinical_Data.groupby(['Drug','Timepoint'])['Mouse ID'].agg('count')

# Convert to DataFrame
miceByDrugTimepoint = pd.DataFrame(miceByDrugTimepoint)
miceByDrugTimepoint.reset_index(inplace = True)
miceByDrugTimepoint.rename(columns = {'Mouse ID': 'Mouse Count'}, inplace = True)
# Preview DataFrame
miceByDrugTimepoint.head()
```

```python
# Minor Data Munging to Re-Format the Data Frames
miceByDrugTimepoint_formatted = miceByDrugTimepoint.pivot(index = 'Timepoint', columns='Drug',values='Mouse Count')

# Preview the Data Frame
miceByDrugTimepoint_formatted.head()
```

```python
# Get the initial count of mice for each drug
miceByDrugTimepoint_formatted_initial = miceByDrugTimepoint_formatted.iloc[0]

#Calculate the survival rate by dividing current live mice / initial mices
miceByDrugTimepoint_formatted_percentage = round(miceByDrugTimepoint_formatted/miceByDrugTimepoint_formatted_initial*100,2)

#Display the survival percentage
miceByDrugTimepoint_formatted_percentage
```

```python
#Get the Timepoint as x_axis_survivalRate
x_axis_survivalRate = miceByDrugTimepoint_formatted_percentage.index.values

# Generate the Plot (Accounting for percentages)

plt.plot(x_axis_survivalRate,miceByDrugTimepoint_formatted_percentage['Capomulin'],label='Capomulin',\
         color ='red', marker = 'o', linestyle ='dashed',linewidth = 0.5, markersize = 5)
plt.plot(x_axis_survivalRate,miceByDrugTimepoint_formatted_percentage['Infubinol'],label='Infubinol',\
         color ='blue', marker = '^', linestyle ='dashed',linewidth = 0.5, markersize = 5)
plt.plot(x_axis_survivalRate,miceByDrugTimepoint_formatted_percentage['Ketapril'],label='Ketapril',\
         color ='green', marker = 's', linestyle ='dashed',linewidth = 0.5, markersize = 5)
plt.plot(x_axis_survivalRate,miceByDrugTimepoint_formatted_percentage['Placebo'],label='Placebo',\
         color ='black', marker = 'd', linestyle ='dashed',linewidth = 0.5, markersize = 5)


plt.legend(loc = 'best')
plt.xlabel('Time (Days)')
plt.ylabel('Survival Rate (%)')
plt.title('Survival During Treatment')
plt.grid()

# Save the Figure

plt.savefig('./Images/3_Survival During Treatment.png')
# Show the Figure
plt.show()
```

## Summary Bar Graph

```python
# Calculate the percent changes for each drug
initial_vol = tumorByDrugTime_Formatted.iloc[0]
final_vol = tumorByDrugTime_Formatted.iloc[len(tumorByDrugTime_Formatted)-1]

pct_chg_vol = (final_vol - initial_vol)/initial_vol*100

# Display the data to confirm
print(pct_chg_vol)
```

```python
tumorByDrugTime_Formatted.head()
```

```python
# Store all Relevant Percent Changes into a Tuple
pctChg_tuple = tuple(pct_chg_vol)

# Splice the data between passing and failing drugs
pct_chg_df = pd.DataFrame(pct_chg_vol)
pct_chg_df.rename(columns = {0:'pct_chg'},inplace=True)
pct_chg_df['Status'] = np.where(pct_chg_df['pct_chg'] >=0,'Fail','Pass')
pct_chg_df['Colors'] = np.where(pct_chg_df['pct_chg'] >=0,'r','g')
pct_chg_df.reset_index(inplace = True)

# Subset the relevant Drugs
pct_chg_df = pct_chg_df.loc[pct_chg_df['Drug'].isin(['Capomulin','Infubinol','Ketapril','Placebo'])]
pct_chg_df.reset_index(inplace = True)
# Get the x axis
x_axis_status = np.array(pct_chg_df['Drug'])

ax = plt.bar(x_axis_status,pct_chg_df['pct_chg'],width=1,align='center',color = pct_chg_df['Colors'] )

# Orient widths. Add labels, tick marks, etc. 

plt.ylabel('% Tumor volume change')
plt.title('Tumor change over 45 days treatment')
plt.grid()
plt.ylim(-28,70)

# Use functions to label the percentages of changes
def get_Percent_changes():
    for i in range(len(x_axis_status)):
        val = pct_chg_df.loc[pct_chg_df['Drug'] == x_axis_status[i], 'pct_chg'].max()
        plt.text(x = x_axis_status[i] , y = pct_chg_df['pct_chg'][i],s = str(round(val,0))+"%",\
                 fontsize = 10, horizontalalignment='center',verticalalignment='bottom',color = 'black')

# Call functions to implement the function calls

get_Percent_changes()

# Save the Figure
plt.savefig('./Images/Tumor Change over 45 days treatment.png')

# Show the Figure
plt.show()
```
